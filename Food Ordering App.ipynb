```python
RESTAURANT_NAME = "Hungry Wolf"

menu = {
    "sku1":{"name": "Hamburger",
            "price": 6.51},
    "sku2":{"name":"CheeseBurger",
            "price":7.75},
    "sku3":{"name":"Milkshake",
            "price":5.99},
    "sku4":{"name":"Fries",
            "price":2.39},
    "sku5":{"name":"Sub",
            "price":5.87},
    "sku6":{"name":"IceCream",
            "price":3.45},
    "sku7":{"name":"Fountain Drink",
            "price":3.45},
    "sku8":{"name":"Cookie",
            "price":3.45},
    "sku9":{"name":"Brownie",
            "price":3.45},
    "sku10":{"name":"Sauce",
            "price":3.45}               
}

app_actions = {
    "1":"Add new menu item to cart",
    "2":"Remove an item from the cart",
    "3":"Modify the cart item's quantity",
    "4":"View cart",
    "5":"Checkout",
    "6":"Exit"
}

SALES_TAX_RATE = 0.07
cart = {}

```

```python
def display_menu():
    print("\n*********MENU************\n")

    for sku in menu:
        parsed_sku = sku[3:]
        item = menu[sku]["name"]
        price = menu[sku]["price"]
        print("(" +parsed_sku+ ")" + " " + item + ": $" + str(price))
display_menu()        
```

```python
def view_cart():
    print("\n*******Cart Menu********\n")
    subtotal = 0
    for sku in cart:
        if sku in menu:
            print (f"You got {cart[sku]} {menu[sku]["name"]} each at price of {menu[sku]["price"]}")
            quantity = cart[sku]
            subtotal += menu[sku]["price"] * quantity
    tax = subtotal * SALES_TAX_RATE
    total = subtotal + tax
    print("Total: $", round(total, 2))
    print("\n")
```

```python
def add_to_cart(sku,quantity=1):
    if sku in menu:
        if sku in cart:
            cart[sku] += quantity
        else:
            cart[sku] = quantity
        print("Added", quantity, " of ", menu[sku]["name"] , "to the cart")
    else:
        print("I'm sorry the menu number",sku, "that you entered is not on the menu")

add_to_cart("sku1", 10)        
```

```python
def remove_from_cart(sku):
    if sku in cart:
        name = menu[sku]["name"]
        removed_val = cart.pop(sku)
        print(name)
        print(f"Removed", {name}, "from the cart. ")
        view_cart()

    else:
        print(f"I'm sorry. The item with SKU {sku} is not currently in the cart")
```

```python
add_to_cart("sku1",10)
remove_from_cart("sku1")
```

```python
def modify_cart(sku,quantity):
    if sku in menu:
        if sku in cart:
            quantity_in_cart = cart[sku]
            if quantity <= 0:
                remove_from_cart(sku)
            else:
                cart[sku] = quantity
        print("Modified", quantity, " of ", menu[sku]["name"] , "to the cart")
    else:
        print("I'm sorry the menu number",sku, "that you entered is not on the menu")
```

```python
def checkout():
    print("******Checkout***********")
    view_cart()
    print("****Thank you for your order**********")
    print("\n")
```

```python
def get_sku_and_quantity(sku_prompt,quantity_prompt=None):
    item_sku = input(sku_prompt)
    item_sku = "sku"+item_sku

    if (quantity_prompt):
        quantity = input(quantity_prompt)

        if not quantity.isdigit():
            quantity = 1
        else:
            quantity = int(quantity)

        return item_sku,quantity
    else:
        return item_sku
    
```

```python
def order_loop():
    print("Welcome to the "+ RESTAURANT_NAME + "!")
    ordering = True
    while ordering:
        print("****Ordering Actions******")
        for number in app_actions:
            description = app_actions[number]
            print("(" + number + ")", description)

        response = input("Please enter the number of the action you want to take:")
        if response == '1':
            display_menu()
            sku_prompt = "Please enter the sku number for the menu item you want to order: "
            quantity_prompt = "Please enter the quantity you want to order(default = 1): "
            ordered_sku,quantity = get_sku_and_quantity(sku_prompt,quantity_prompt)
            add_to_cart(ordered_sku,quantity)
        elif response == "2":
            display_menu()
            sku_prompt = "Please enter the sku number for the menu item you want to remove: "
            item_sku = get_sku_and_quantity(sku_prompt)
            remove_from_cart(item_sku)
        elif response == "3":
            display_menu()
            sku_prompt = "Please enter the sku number for the menu item you want to modify: "
            quantity_prompt = "Please enter the quantity you want to modify to(default = 1): "
            item_sku,quantity = get_sku_and_quantity(sku_prompt,quantity_prompt)
            modify_cart(item_sku,quantity)
        elif response == "4":
            view_cart()
        elif response == "5":
            checkout()
            ordering = False
        elif response == "6":
            print("Goodbye!")
            ordering = False
        else:
            print("You entered invalid action number. Please try again.")






```

```python
order_loop()
```
